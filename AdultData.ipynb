```python
#Marital Status in more general terms
adultdata.replace(['Divorced', 'Married-AF-spouse', 
              'Married-civ-spouse', 'Married-spouse-absent', 
              'Never-married','Separated','Widowed'],
             ['not married','married','married','married',
              'not married','not married','not married'], inplace = True)
```

```python
import numpy as np
for col in category_col:
    b, c = np.unique(adultdata[col], return_inverse=True) 
    adultdata[col] = c

adultdata.head()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
array = adultdata.values
X = array[:,0:8]
Y = array[:,8]
rfe = RFE(model)
fit = rfe.fit(X, Y)
i=0
feature=list()
print('Num Features:',fit.n_features_)
for x,y in zip(fit.support_,adultdata):
    if x:
        feature.append(y)
print("Feature Ranking:",fit.ranking_)
print("Selected Features:",feature)
```
